# operator chaining

    python supports chaining like a<b<c. 
    
    chaining is performed from left to right.
    
    (a<b) is checked first, if true (b<c) is evaluated else not
    
    (a<c) is not checked

```python
age=float(input('enter your age: '))
if 17 < age <= 23:
    print('you might be a undergrad student')
else:
    print('when did you complete your b.tech degree')
```

```python
# can use the logical operators 'and', 'or', 'not'

age=float(input('enter your age: '))
workexp=float(input('enter your work experience in years: ')) 
if (age > 17) and (age <= 23):
    print('you might be a undergrad student')
elif (age > 23) and (age < 28) and (0 < workexp < 5):
    print('when did you complete your b.tech degree')
elif (age > 30) or (not (workexp < 5)):
    print('have you done masters?')
else:
    print('what is your educational background?')
    
```

```python
# can do better if a nested if-else is use
age=float(input('enter your age: '))
workexp=int(input('enter your work experience in years: '))
if workexp == 0:
    if (17 <= age <=23):
        print('you might be a undergrad student')
    elif (age < 26):
        print('when did you complete your b.tech degree')
    else:
        print("why don't you do a masters?")
else:
    print('what organization do you work for?')
```

# comparing data types using relational operators

    avoid comparing floating point numbers as they may be stored differently
    
    strings are compared litral by literal. generally it is converted to (ASCII     or unicode) and then compared
    
    lists and tuples are compared by their values in each index
    
    dictionaries are compared using == and != only. they are compared through
    the key:value pairs.

```python
# float comarison
x=1/3
y=0.333
if (x == y):
    print('you are wrong')
else:
    print('its interesting')
```

```python
# string comparison
today='monday'
if today != 'Monday':
    print('thanks god i do not have to wake up early for python class')
else:
    print('oh no. i have to wake up early')
```
