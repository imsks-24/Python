```python
# copying particular contents from one file to another
oname = input("Output file name: ")
f = int(input("From line: "))
t = int(input("To line: "))
# Open both files
with open("txtandotherfiles_for_pythoncourse/lec9_ex2", mode="rt", encoding="utf8") as ifile, \
     open(oname, mode="wt+", encoding="utf8") as ofile:
    # if you want to Copy the first line uncomment and change the range() args
    #ofile.write(ifile.readline())
    # Read the first `f-1` lines, as we don't want to copy these
    for _ in range(f-1):
        if ifile.readline() == "":
            break
    # Copy the following `t-f+1` lines:
    for k in range(t-f+1):
        line = ifile.readline()
        if line == "":
            break
        ofile.write(line)
```

```python
# a file contains city names and ages of persons living there
# take a user input for city name and tell the average age of the city

with open("txtandotherfiles_for_pythoncourse/lec9_ex5", mode="rt", encoding="utf8") as ifile:
    lt2=ifile.readlines()

city=input('enter the city name')
for item in lt2:
    if city in item:
        lt3=(item.strip(city)).split()
        lt4=[int(val) for val in lt3]
        print(sum(lt4)/len(lt4))
     
```
